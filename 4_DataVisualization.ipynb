## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# Fetch all Age and WorkWeekHrs values
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

df = pd.read_sql_query(QUERY, conn)

# Plot histogram
df.plot(kind='scatter', x='Age', y='WorkWeekHrs', alpha = 0.7, color='purple')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# Fetch all Age, CodeRevHrs WorkWeekHrs values
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""

df = pd.read_sql_query(QUERY, conn)

# Plot histogram
ax = df.plot(kind='scatter', x='CodeRevHrs', y='WorkWeekHrs', s=df['Age'], alpha=0.7, edgecolors="w", linewidth=0.5, color ='purple')

plt.title('Bubble Plot: Age vs. WorkWeekHrs (Bubble Size by Salary)')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# Fetch all values
QUERY = """
SELECT DatabaseDesireNextYear
FROM DatabaseDesireNextYear
"""

df = pd.read_sql_query(QUERY, conn)

# TOP 5 DB TOP 5 COUNT
database_count = df.value_counts()
top5database = database_count.head(5)

print(df.value_counts())
# Plot histogram
top5database.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Top Five Databases')
plt.ylabel('') 
plt.show()
```

```python
# Fetch all values
QUERY = """
SELECT LanguageWorkedWith
FROM LanguageWorkedWith
"""

df = pd.read_sql_query(QUERY, conn)

# TOP 5 DB TOP 5 COUNT
language_count = df.value_counts()
top5languages = language_count.head(5)

print(df.nunique())

# Plot histogram
top5languages.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Top Five Languages Worked With')
plt.ylabel('') 
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# Fetch all values
QUERY = """
SELECT ConvertedComp, Age
FROM Master
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY, conn)

#Create a median variable for the ConvertedComp by age (45-60)
mask = (df['Age'] >= 45) & (df['Age'] <= 60)


converted_comp_values = df[mask]['ConvertedComp']
median_comp = converted_comp_values.median()

median_by_age = df[mask].groupby('Age')['ConvertedComp'].median()

median_by_age.plot()
plt.title('Median Converted Compensation by Age [45 - 60]')
plt.ylabel('Median ConvertedComp')
plt.show()
```
