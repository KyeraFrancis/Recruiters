```python
# Define the 'arts' and 'non-arts' categories
arts_categories = ["Art", "Dance", "Music", "Photography", "Theater", "Film & Video"]

# Create a new column in the dataframe to group categories as 'arts' or 'non-arts'
df['grouped_category'] = df['category'].apply(lambda x: 'Arts' if x in arts_categories else 'Non-Arts')

# Filter out campaigns based on 'successful' status
successful_campaigns = df[df['status'] == 'successful']

# Count the total number of campaigns per grouped category
total_campaigns = df['grouped_category'].value_counts()

# Count the number of successful campaigns per grouped category
successful_counts = successful_campaigns['grouped_category'].value_counts()

# Calculate success rate for each grouped category
success_rate = (successful_counts / total_campaigns) * 100

# Plot
plt.figure(figsize=(8, 5))
success_rate.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Success Rate by Grouped Category')
plt.ylabel('Success Rate (%)')
plt.xlabel('Grouped Category')
plt.xticks(rotation=0)
plt.text(-0.05, 55, '58%')
plt.text(.9705,35,'42%')
plt.tight_layout()
plt.show()
```

```python
# Set the style of the visualization
sns.set(style="whitegrid")


# Filter successful and failed campaigns
successful_campaigns = df[df['status'] == 'successful']
failed_campaigns = df[df['status'] == 'failed']

# Plot
plt.figure(figsize=(12, 8))

# Histograms and KDEs
sns.histplot(successful_campaigns['duration'], kde=True, color="green", label="Successful", bins=30)
sns.histplot(failed_campaigns['duration'], kde=True, color="red", label="Failed", bins=30)

# Titles and labels
plt.title('Distribution of Campaign Duration for Successful vs. Failed Campaigns')
plt.xlabel('Duration (days)')
plt.ylabel('Number of Campaigns')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Calculate the average goal for successful and failed campaigns
average_successful_goal = df[df['status'] == 'successful']['goal'].mean()
average_failed_goal = df[df['status'] == 'failed']['goal'].mean()

# Data for plotting
categories = ['Successful', 'Failed']
values = [average_successful_goal, average_failed_goal]

# Plotting
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=['green', 'red'])
plt.title('Average Campaign Goal for Successful vs. Failed Campaigns')
plt.ylabel('Average Goal ($)')
plt.xlabel('Campaign Status')
plt.show()




```

```python
top_campaigns = df.sort_values(by='funded percentage', ascending=False).head(10)

# Plotting
plt.figure(figsize=(15, 8))
plt.barh(top_campaigns['category'], top_campaigns['funded percentage'], color='dodgerblue')
plt.xlabel('Funded Percentage')
plt.ylabel('Campaign Category')
plt.title('Top 10 Most Successfully Funded Campaign Type ')
plt.gca().invert_yaxis()  # This will have the highest funded percentage at the top
plt.tight_layout()
plt.show()
```

```python
# Convert 'funded date' to datetime
df['funded date'] = pd.to_datetime(df['funded date'])

# Extract month from the 'funded date'
df['month'] = df['funded date'].dt.month

# Group by month and calculate success rate
success_rate_by_month = df[df['status'] == 'successful'].groupby('month').size() / df.groupby('month').size()

# Define a specific color for each month
month_colors = [
    'grey', 'grey', 'green', 'green', 'grey', 
    'red', 'red', 'grey', 'grey', 'grey',
    'green', 'red'
]
# Plotting
plt.figure(figsize=(10, 6))
success_rate_by_month.plot(kind='bar', color=month_colors)
plt.title('Kickstarter Success Rate by Month')
plt.xlabel('Month')
plt.ylabel('Success Rate')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.tight_layout()
plt.show()
```
